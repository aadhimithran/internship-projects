```python
data = {
    'ID': [1, 2, 3, 4, 5],
    'Age': [51, 31, 32, 38, 32],
    'Attrition': [0, 1, 0, 0, 0],
    'DistanceFromHome': [6, 10, 17, 2, 10],
    'Education': [2, 1, 4, 5, 1],
    'EmployeeCount': [1, 1, 1, 1, 1],
    'JobLevel': [1, 1, 1, 1, 1],
    'JobRole': ['Healthcare Representative', 'Research Scientist', 'Sales Executive', 'Human Resources', 'Sales Executive'],
    'MaritalStatus': ['Married', 'Single', 'Married', 'Married', 'Single'],
    'MonthlyIncome': [131160, 41890, 193280, 83210, 23420]
}
```

```python
job_role_mapping = {
    'Healthcare Representative': 1,
    'Research Scientist': 2,
    'Sales Executive': 3,
    'Human Resources': 4,
    'Laboratory Technician': 5,
    'Manager': 6,
    'Manufacturing Director': 7,
    'Sales Representative': 8,
    'Research Director': 9
}

# Map the 'JobRole' column to numerical values
dataset['JobRole'] = dataset['JobRole'].map(job_role_mapping)


print(dataset[['JobRole']])
```

```python
plt.figure(figsize=(15,5))
plt.subplot(2, 3, 1)
sns.histplot(dataset['Age'], bins=10, kde=True)
plt.title('Age Distribution')
plt.show()
```

# Attrition by Age

```python
plt.figure(figsize=(15,5))
plt.subplot(2, 3, 2)
sns.countplot(x='Age', hue='Attrition', data=dataset)
plt.title('Attrition by Age')
plt.show()
```

# Monthly Income Distribution

```python
plt.figure(figsize=(10,10))
plt.subplot(2, 3, 3)
sns.boxplot(x='MonthlyIncome', data=dataset)
plt.title('Monthly Income Distribution')
plt.show()
```

# Job Role Distribution

```python

plt.figure(figsize=(15,10))
plt.subplot(2, 3, 4)                                            
sns.countplot(x='JobRole', data=dataset)
plt.title('Job Role Distribution')
plt.xticks(rotation=45)
plt.show()

#'Healthcare Representative': 1,
#'Research Scientist': 2,
#'Sales Executive': 3,
#'Human Resources': 4
            
```

# Attrition by Education Level

```python

plt.figure(figsize=(4, 4))
sns.countplot(x='Education', hue='Attrition', data=dataset)
plt.title('Attrition by Education Level')
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.subplot(2, 3, 6)
sns.heatmap(dataset.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
